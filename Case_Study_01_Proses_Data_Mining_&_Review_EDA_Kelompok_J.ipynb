# **Rumah dengan karakteristik seperti apa yang paling banyak dijual?**

```python
gabung["KT"] = gabung["KT"].astype("category")
gabung["KM"] = gabung["KM"].astype("category")
gabung["listrik"] =gabung["listrik"].astype("category")
gabung["carport"] =gabung["carport"].astype("category")
```

```python
# Definisikan fungsi untuk mengganti outlier dengan "lainnya"
def replace_outliers_with_lainnya(data, column, threshold=0.05, additional_categories=['PPJB', 'GIRIK', 'ADAT']):
  outlier_categories = data[column].value_counts()[data[column].value_counts() < threshold * len(data)].index
  all_other_categories = outlier_categories.union(additional_categories)  # Gabungkan dengan kategori tambahan
  data[column] = data[column].apply(lambda x: 'Lainnya' if x in all_other_categories else x)
  return data

# Ganti outlier dengan "lainnya" untuk setiap variabel, termasuk 'sertifikat' dengan kategori tambahan
gabung2 = replace_outliers_with_lainnya(gabung, 'KT')
gabung2 = replace_outliers_with_lainnya(gabung, 'KM')
gabung2 = replace_outliers_with_lainnya(gabung, 'listrik')
gabung2 = replace_outliers_with_lainnya(gabung, 'carport')
gabung2 = replace_outliers_with_lainnya(gabung, 'sertifikat', additional_categories=['PPJB', 'GIRIK', 'ADAT'])

# Visualisasi subplot
fig, axes = plt.subplots(3, 2, figsize=(15, 15))  # Ubah ukuran subplot

# Subplot 1: KT
sns.countplot(x='KT', data=gabung2, ax=axes[0, 0], order=gabung['KT'].value_counts().index)
axes[0, 0].set_title('Banyaknya Jumlah Kamar Tidur pada Rumah yang Dijual')

# Subplot 2: KM
sns.countplot(x='KM', data=gabung2, ax=axes[0, 1], order=gabung['KM'].value_counts().index)
axes[0, 1].set_title('Banyaknya Jumlah Kamar Mandi pada Rumah yang Dijual')

# Subplot 3: listrik
sns.countplot(x='listrik', data=gabung2, ax=axes[1, 0], order=gabung['listrik'].value_counts().index)
axes[1, 0].set_title('Sebaran Kategori "listrik"')
axes[1, 0].tick_params(axis='x', rotation=45)

# Subplot 4: carport
sns.countplot(x='carport', data=gabung2, ax=axes[1, 1], order=gabung['carport'].value_counts().index)
axes[1, 1].set_title('Sebaran Kategori "carport"')

# Subplot 5: sertifikat
sns.countplot(x='sertifikat', data=gabung2, ax=axes[2, 0], order=gabung['sertifikat'].value_counts().index)  # Tambahkan subplot 'sertifikat'
axes[2, 0].set_title('Sebaran Kategori "sertifikat"')
axes[2, 0].tick_params(axis='x', rotation=45)

# Hapus subplot kosong (jika ada)
fig.delaxes(axes[2, 1])

plt.tight_layout()
plt.show()
```

```python
#Buang outliers dari Luas Bangunan
Q1 = gabung['LB'].quantile(0.25)
Q3 = gabung['LB'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
gabung2 = gabung[(gabung['LB'] >= lower_bound) & (gabung['LB'] <= upper_bound)]

#membuat histogram banyaknya per rentang luas bangunan
plt.figure(figsize=(10, 6))
plt.hist(gabung2['LB'], bins=10, edgecolor='black')
plt.title('Histogram Luas Bangunan')
plt.xlabel('Luas Bangunan')
plt.ylabel('Frekuensi')
plt.show()
```

Dari histogram penjualan rumah di beberapa kota besar seperti Jakarta, Depok, Bekasi, Tangerang, dan Bogor, kesimpulan utama terkait rumah yang paling banyak dijual adalah rumah dengan karakteristik:

* 2 kamar tidur dan 2 kamar mandi â€“ menunjukkan preferensi pasar terhadap rumah ukuran kecil hingga menengah yang cukup untuk keluarga kecil.
* Kapasitas listrik 2200 watt â€“ ini mencerminkan rumah yang sesuai dengan kebutuhan konsumsi listrik untuk keluarga modern.
* Sertifikat Hak Milik (SHM) â€“ menunjukkan bahwa pembeli lebih memilih properti dengan kepastian hak milik.

* Luas bangunan antara 50-150 mÂ² â€“ mengindikasikan bahwa rumah ukuran menengah lebih diminati dibandingkan rumah kecil atau besar.

Sehingga dapat disimpulkan berdasarkan data, rumah yang banyak dijual pada saat ini adalah rumah dengan ukuran menengah dan memiliki fasilitas standar yang berada pada daerah bekasi, bogor, dan tangerang karena memiliki harga rumah yg relatif stabil dan cocok untuk minat pasar pada saat ini
